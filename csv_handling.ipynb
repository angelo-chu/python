```python
import pandas as pd

# Define the paths to the CSV files
file1 = '/content/drive/My Drive/Coding/housing.csv'
file2 = '/content/drive/My Drive/Coding/housing2.csv'

# Read the CSV files into DataFrames
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Concatenate the DataFrames
df = pd.concat([df1, df2], ignore_index=True)

# Print the DataFrame
df.tail()

df.to_csv('/content/drive/My Drive/Coding/merged_housing.csv', index=False)
df3 = pd.read_csv('/content/drive/My Drive/Coding/merged_housing.csv')
df3.tail()
```

# Clear a column in multiple csv files


```python
import pandas as pd
import glob
import os

# Mount Google Drive (if your files are on Drive)
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# Set the working directory to where your files are located
os.chdir('/content/drive/My Drive/Coding/Clear column')

# Get a list of all CSV files in the directory
file_list = glob.glob("*.csv")
file_list

# Process each file
for filename in file_list:
    df = pd.read_csv(filename)

    # Clear the values in the "Customer external address" column
    df["Address"] = ""

    # Save the modified DataFrame back to the same CSV file
    df.to_csv(filename, index=False)

print("address cleared in all files!")
```
