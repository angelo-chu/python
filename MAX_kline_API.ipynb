```python
import requests
import pandas as pd
```

```python
API_URL = "https://max-api.maicoin.com/api/v2/k"  # Replace with actual API URL
market = "btcusdt"  # Specify the market (e.g., "ethusdt")
limit = 412
period = 1440  # Set period to 4320 for 1 month
timestamp = 1661798400  # Converted Unix epoch for September 1st, 2022

# Build the query string with parameters
params = {
    "market": market,
    "limit": limit,
    "period": period,
    "timestamp": timestamp,
}
```

```python
response = requests.get(API_URL, params=params)
```

```python
if response.status_code == 200:
  # Request successful
  data = pd.DataFrame(response.json())
  print(data)
  # Process the data (array of timestamps, open, high, low, close, volume)
else:
  print(f"Error: API request failed with status code {response.status_code}")
```

```python
#convert unix epoch to date and time of UTC + 8

from datetime import datetime, timezone, timedelta
```

```python
def unix_to_utc8(timestamp):
  """Converts a Unix timestamp to a UTC+8 datetime object."""
  # Convert timestamp to datetime object in UTC
  dt = datetime.utcfromtimestamp(timestamp)
  # Convert to UTC+8 timezone
  return dt + timedelta(hours=8)
```

```python
# Assuming your DataFrame is called 'data' with a column named 'timestamp'
data['datetime_utc8'] = data[0].apply(unix_to_utc8)
```
