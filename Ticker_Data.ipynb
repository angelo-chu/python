```python
# @title Set stock symbol

symbol = 'VOO' # @param {type:"string"}

print('system has recorded： '+ symbol)
```

```python
# @title Import Library

import datetime
!pip install pandas_datareader --upgrade
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override() #referencing https://github.com/ranaroussi/yfinance?tab=readme-ov-file#pandas_datareader-override

from pandas_datareader import get_data_yahoo

import plotly.express as px
import plotly.graph_objects as go
```

```python
# @title Pull ticker data
noOfYears = 5
endDate = datetime.datetime.now()
startDate = endDate - datetime.timedelta(days = 365 * noOfYears)

print('Extracting data from  ', startDate, ' to ', endDate, ' \n')

data = pdr.get_data_yahoo(symbol, startDate, endDate)
print('Extraction successful! Data are as follows: \n', data)
```

```python
# @title Visualization

data = data.reset_index()

fig = px.line(data, x= "Date", y = "Adj Close", title = symbol + ' price trend')
fig.show()
```
